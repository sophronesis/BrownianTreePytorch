## Brownian Tree with numpy:

```python
import numpy as np
import matplotlib.pylab as plt
from tqdm import tqdm
import utils as ut
```

```python
# Constants
gridSize = 301 # grid size (odd)
assert(gridSize % 2 == 1)
percParticles = 10
partNum = gridSize * gridSize * percParticles // 100 # moving particles
iters = 1500
```

```python
# Initialisation of fields
ffield, mfield = [np.zeros((gridSize, gridSize), dtype=np.uint8) for i in range(2)]
ffield[gridSize//2, gridSize//2] = 1
ut.placeDots(mfield, partNum)

# Main cycle
for i in tqdm(range(iters)):
    ffield, mfield = ut.freezeDots(ffield, mfield)
    mfield = ut.moveDots(mfield)
```

```python
temp = [2 * ffield + mfield]
fig, axs = plt.subplots(1, len(temp), figsize=(20, 20))
for n, i in enumerate(temp):
    axs.imshow(i, cmap=plt.get_cmap("gray"))
plt.show()
```
